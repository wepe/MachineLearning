```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
number_of_points = 10
X_values = np.matrix([i for i in range(number_of_points)]).T
X = X_values[:]
print(X)
m,n = np.shape(X)
```

```python
y_values = np.matrix([0.4*i+3+6*np.random.random() for i in range(number_of_points)])
y = y_values[:]
y
```

Let m be the number of training examples.
And n be the number of features in the initial setup.

Here, in this example m = 10 and n = 1.
The X matrix is a matrix of order mxn.
The y matrix is a matrix of order 1xm

The goal, firstly, is to make the X matrix in the design matrix form of the order (n+1)xm

```python
X = np.concatenate([np.matrix(np.ones(m)), X.T])
```

```python
theta = np.matrix(np.zeros(n+1)).T
```

```python
J = 0
learning_rate = 0.001
prev_J = 1
number_of_iterations = 0
while abs(prev_J-J)>0:
    number_of_iterations+=1
    prev_J = J
    J = (1/(2*m))*np.sum(np.power(y-theta.T@X, 2))
    derivative = (1/m)*(((y-theta.T@X)@X.T).T)
    theta = theta - learning_rate*derivative
    print("J = {}\tm = {}, c = {}".format(J, theta.item(1), theta.item(0)))

theta
```
